Accesing the RGB values for the given index

```python

#changing pixel values

from PIL import Image
import numpy as np
im = np.array(Image.open("lena_square.png"))

print(im.shape)

print(im[100,150])

print(type(im[100,150]))
```

```python
#unpacking a tupe and list in python for RGB space

R,G,B = im[100,150]

print(R)

print(G)

print(B)
```

```python
#get the value by specifying the color space

print(im[100,150,0])

print(im[100,150,1])

print(im[100,150,2])
print(type)
```

```python
print(im[100,150].shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(im)
```

```python
#Convert the given color image into black and white by using THreshholding concept
for i in range(0,512):
    for j in range(0,512):
        for c  in range(3):
            if im[i,j,c]> 128:
                im[i,j,c] = 150
            else:
                im[i,j,c] = 250
```
