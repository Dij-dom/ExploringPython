## Data munging
#### Also known as data wrangling, is the process of cleaning, transforming, and enriching raw data into a more structured and usable format

```python
data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'FirstName': ['John', 'Alice', 'Bob', 'Eva', 'David'],
    'LastName': ['Doe', 'Smith', 'Johnson', 'Lee', 'Brown'],
    'Salary': ['$60,000', '$55,000', '$70,000', '$62,000', '$58,000'],
    'HireDate': ['2022-01-15', '2021-03-10', '2022-05-20', '2020-12-05', '2022-08-30']
}

dm = pd.DataFrame(data)
dm
```

```python
# Data Munging Steps


# 1. Convert the 'HireDate' column to a datetime object
dm['HireDate'] = pd.to_datetime(dm['HireDate'])

# 2. Create a new column 'YearsSinceHire' based on the 'HireDate'
dm['YearsSinceHire'] = pd.Timestamp.now().year - dm['HireDate'].dt.year

# 3. Sort the DataFrame by 'Salary' in descending order
dm = dm.sort_values(by='Salary', ascending=False)

# 4. Reset the index of the DataFrame
dm = dm.reset_index(drop=True)

# Display the munged DataFrame
print("\nMunged DataFrame:")
print(dm)
```
